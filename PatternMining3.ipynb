```python
transaction = []
for i in range(0, titanic.shape[0]):
    for j in range(0, titanic.shape[1]):
        transaction.append(titanic.values[i,j])

transaction = np.array(transaction)
type(transaction)
```

```python
df = pd.DataFrame(transaction, columns = ['items'])
df['incident_count'] = 1
df
```

```python
indexNames = df[df['items'] == 'nan'].index
df.drop(indexNames, inplace=True)

df_table = df.groupby('items').sum().sort_values('incident_count', ascending=False).reset_index()
print(df_table)
```

```python
transaction2 = []
for i in range(titanic.shape[0]):
    transaction2.append([str(titanic.values[i,j]) for j in range(titanic.shape[1])])
```

```python
type(transaction2)
```

```python
transaction2
```

```python
transaction2 = np.array(transaction2)
```

```python
transaction2
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction2).transform(transaction2)
```

```python
te_ary
```

```python
titanic2 = pd.DataFrame(te_ary, columns = te.columns_)
titanic2
```

```python
first35 = df_table['items'].head(35).values
titanic3 = titanic2.loc[:,first30]
titanic3.shape
```

```python
res=fpgrowth(titanic3, min_support=0.1, use_colnames=True)
res.head(35)
```

```python
rules = association_rules(res, metric = 'confidence', min_threshold = 0.7)
rules
```

```python
filtered_rules = rules[(rules['consequents'] == {'Yes'}) | (rules['consequents'] == {'No'})]
filtered_rules
```
